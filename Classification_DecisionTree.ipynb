* 데이터에 있는 규칙을 학습을 통해 자동으로 찾아내 트리 기반의 분류 규칙을 만드는 것 <br>
* 규칙 (노드) 이 많다는 것은 그만큼 복잡하다는 것이고 이는 '과적합' 으로 이어질 수 있기 때문에 depth 가 깊어질수록 성능이 저하<br>

즉, 높은 예측 정확도를 가지려면 데이터를 분류 할 때 최대한 많은 데이터 세트가 해당 분류에 속할 수 있도록 규칙 노드가 정해져야 한다.<br>그럼 규칙 조건이 정해지는 방식을 알아보자.<br><br>
규칙 조건은 <u>균일도</u> 가 높은 데이터 세트를 먼저 분할 할 수 있도록 조건이 정해진다. <br>
> Example <br><br>
쉽게 말해, 검은 색 바둑알로 가득찬 A 상자 , 검은 색과 흰 색 바둑알이 8 : 2 으로 찬 B 상자 , 검은 색과 흰 색 바둑알이 5 : 5 로 찬 C 상자가 있다. 균일도가 높은 순서는 A > B > C 이고, 위 규칙 조건 원리에 의해 첫 번 째 규칙 조건은 ' 바둑알 색 == 검은색 ' 이 될 것이다. 

이러한 데이터의 균일도를 측정하는 방법은 <br>
* __엔트로피를 이용한 정보 이득 지수__
    * 엔트로피는 주어진 데이터 집합의 혼잡도를 의미. 즉, 서로 다른 값이 섞여 있으면 엔트로피가 높고, 같은 값이 섞여 있으면 엔트로피가 낮음
    * 정보 이득 지수 = 1 - 엔트로피 
    * 결국, 정보 이득 지수 ↑ =  균일도 ↑ = 정보 이득이 높은 속성을 먼저 분할 
    <br><br>
* __지니 계수__ 
    * 머신러닝에서는, 지니 계수 ↓ = 균일도 ↑ =  지니 계수가 낮은 속성을 먼저 분할 

#### 결론 
결정 트리는 데이터 세트를 분할 하는데 정보 이득을 높이거나 지니 계수가 낮은 조건을 찾아서 자식 트리 노드에 걸쳐 반복적으로 분할 한 뒤, 데이터가 모두 특정 분류에 속하게 되면 분할을 멈추고 분류를 결정한다.

#### 장점
feature scaling 등의 사전 가공 영향을 크게 받지 않고 '균일도' 만 신경 쓰면 되서 직관적

### 결정 트리 모델의 시각화 

어떠한 규칙을 가지고 트리를 생성하는지 시각적으로 확인하기 위해  Graphviz  패키지를 사용한다.<br>
[상세 정보 페이지](http://graphviz.org/)<br><br>
앞선 장에서 계속 사용했던 내장 dataset iris 를 가져와서 학습 시킨 뒤 어떠한 형태로 규칙 트리가 만들어지는지 시각화 해보자.<br><br>
사이킷 런의 트리 모듈은 Graphviz 를 이용하기 위해, export_graphviz( ) 함수를 사용한다.
*[export_graphviz( ) Guide](https://scikit-learn.org/stable/modules/generated/sklearn.tree.export_graphviz.html)*<br><br>
해당 함수를 사용하면 Graphviz 에서 파라미터들을 읽어 들여서 그래프 형태로 시각화 할 수 있는 .dot 출력 파일을 생성해준다.<br>
그 후 생성된 파일을 Graphviz 가 Source() 를 통해 읽어서 주피터 노트북 상에서 시각화 시켜준다.


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import export_graphviz
import graphviz
import warnings
warnings.filterwarnings('ignore')

dt_clf = DecisionTreeClassifier(random_state=156)

iris_data = load_iris()
X_train , X_test, y_train , y_test = train_test_split(iris_data.data,iris_data.target,test_size=0.2,random_state=11)

dt_clf.fit(X_train,y_train)


export_graphviz(dt_clf,out_file="tree.dot",class_names=iris_data.target_names,feature_names=iris_data.feature_names,impurity=True,filled=True)

with open("tree.dot") as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

leaf node 는 레이블 값이 결정된 노드이고, branch node 는 자식 노드를 만들기 위한 분할 규칙 조건을 가지고 있는 노드 이다.<br><br>
각 노드의 구성 요소를 살펴보자<br>

* gini : value=[ ] 로 주어진 데이터 분포에서 지니 계수
* feature name <= 값 : 규칙 조건 
* samples : 현 규칙에 해당하는 데이터 건수 
* class = value 중 건 수가 많은 label name
* value = [ ] : 레이블 값 기반의 데이터 건수 , 다 합치면 samples 수
    
    ex) iris dataset 에서 label 은 0:Setosa, 1:Versicolor, 2:Virginica 이므로 <br>   value = [41,40,39] 라면 label 0 이 41개 label 1 이 40개 label 2 가 39 개 의 데이터로 구성됨을 의미
    
#### <span style="color:red">이해를 위한 정리 </span>
앞서 상자 예제에서 균일도가 높은 경우는 상자 안에 한 가지 label 비율이 높을 때 였고, 여러 가지 label 이 골고루 있을 수록 균일도가 낮았다. <br>이를 생각해 보면, 위 결정 트리 Graphviz 에서 value = [   ] 에 한 가지 label 만 의 건 수가 많을 수록 균일도가 높은 셈이 된다. <br> 이 때, gini 계수는 균일도와 반비례 하므로 여러 label 들이 골고루 있을 수록 gini 계수는 높아지고 한 label 로 구성 될 수록 gini 계수는 낮아진다. <br> 아마(?) 균일도 = 건 수 라 봐도 무방하고, 균일도가 높은 데이터를 먼저 분할 할 수 있도록 규칙 조건이 정해진 다 했으므로 Root node 를 보면, label 0 번이 먼저 분할 된 것을 확인 할 수 있다.
